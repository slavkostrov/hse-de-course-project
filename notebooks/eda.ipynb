```python
import pandas as pd

DATA_PATH = globals()["_dh"][0].parent / "data"
NA_VALUES = [""]
```

```python
passport_blacklist_df = pd.read_excel(DATA_PATH / "passport_blacklist_01032021.xlsx", na_values=NA_VALUES)
terminals_df = pd.read_excel(DATA_PATH / "terminals_01032021.xlsx", na_values=NA_VALUES)
transactions_df = pd.read_csv(DATA_PATH / "transactions_01032021.txt", delimiter=";", na_values=NA_VALUES)
```

```python
for df in passport_blacklist_df, terminals_df, transactions_df:
    display(df.head(5))
    display(df.isna().sum().to_frame(name="Количество пропусков"))
    for column in df.columns:
        print("-" * 30, column, "-" * 30)
        n_unique_values = df[column].nunique()
        print("nunique_values:", n_unique_values)
        if n_unique_values < 10:
            print(df[column].unique())
```
