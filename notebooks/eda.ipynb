```python
passport_blacklist_df = pd.read_excel(DATA_PATH / "passport_blacklist_01032021.xlsx", na_values=NA_VALUES)
terminals_df = pd.read_excel(DATA_PATH / "terminals_01032021.xlsx", na_values=NA_VALUES)
transactions_df = pd.read_csv(DATA_PATH / "transactions_01032021.txt", delimiter=";", na_values=NA_VALUES, decimal=",")

clients_df = pd.read_sql("select * from info.clients", con=connection)
cards_df = pd.read_sql("select * from info.cards", con=connection)
accounts_df = pd.read_sql("select * from info.accounts", con=connection)
```

```python
passport_blacklist_full_df = []
terminals_full_df = []
transactions_full_df = []
for name in "01032021", "02032021", "03032021":
    passport_blacklist_df = pd.read_excel(DATA_PATH / f"passport_blacklist_{name}.xlsx", na_values=NA_VALUES)
    terminals_df = pd.read_excel(DATA_PATH / f"terminals_{name}.xlsx", na_values=NA_VALUES)
    transactions_df = pd.read_csv(DATA_PATH / f"transactions_{name}.txt", delimiter=";", na_values=NA_VALUES, decimal=",")

    passport_blacklist_full_df.append(passport_blacklist_df.assign(name=name))
    terminals_full_df.append(terminals_df.assign(name=name))
    transactions_full_df.append(transactions_df.assign(name=name))

passport_blacklist_full_df = pd.concat(passport_blacklist_full_df)
terminals_full_df = pd.concat(terminals_full_df)
transactions_full_df = pd.concat(transactions_full_df)
```

```python
full_df = (
    transactions_full_df
    .merge(terminals_full_df.drop_duplicates(subset=["terminal_id"]), left_on="terminal", right_on="terminal_id")
    .merge(cards_df.assign(card_num=lambda df: df["card_num"].apply(str.strip)), on="card_num")
    .merge(accounts_df, on="account")
    .merge(clients_df, left_on="client", right_on="client_id")
    .merge(passport_blacklist_full_df[["passport"]].assign(blacklist=True).drop_duplicates(), left_on="passport_num", right_on="passport", how="left")
)
```
